```python
class_names = ['T-shirt/top', 'Trauser', 'Pullover', 'Dress',
               'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
class DatasetTypeUnknoun(Exception):
    pass

def show_image(img, label, ax=None, figsize=None):
    if ax is None: fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(img)
    ax.set_title(label)
    
def show_index(index, dataset:str='train', ax=None, figsize:(int, int)=(5, 5)):
    if dataset == 'train'  : img, label = x_train[index], y_train[index]
    elif dataset == 'valid': img, label = x_test[index], y_test[index]
    else: raise DatasetTypeUnknoun 
    label = class_names[label]
    show_image(img, label, ax, figsize)

def show_batch(nrows, ncols, dataset:str='train', figsize:(int, int)=(5, 5)):
    fig, axs = plt.subplots(nrows, ncols, figsize=(figsize[0]*nrows, figsize[1]*ncols))
    if dataset == 'valid': 
        ids = np.random.permutation(len(x_test))[:nrows*ncols].reshape((nrows, ncols))
    else:
        ids = np.random.permutation(len(x_train))[:nrows*ncols].reshape((nrows, ncols))
    for i in range(nrows):
        for j in range(ncols):
            show_index(ids[i,j], dataset, axs[i, j])
    

show_batch(3, 3, 'valid')
```

# Simple Model (Dense)

```python
# Building the model
def build_dense(input_shape, layers_=[256, 128], n_classes=10, drop=0.2, name='Dense'):
    model = keras.Sequential(name=name)
    model.add(layers.Flatten(input_shape = input_shape))
    
    for l in layers_:
        model.add(layers.Dense(l, activation='relu'))
        model.add(layers.Dropout(drop))
        
    model.add(layers.Dense(n_classes, activation='softmax'))
    return model

model = build_dense((28, 28), [256, 128], 10)
```

```python
# Compiling the model
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=["accuracy"])
```

```python
model.summary()
```

```python
# Training the model
H = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=32, epochs=5)
```

```python
def plot_curve(history, metrics=['accuracy', 'loss'], valid:bool=False):
    history = history.history
    for metric in metrics:
        plot_metric(history, metric, valid)
            
def plot_metric(history, metric, valid, ax=None):
    if ax is None:
        _, ax = plt.subplots(figsize=(5, 5))
        ax.plot(history[metric])
        if valid:
            ax.plot(history['val_' + metric])
        ax.set_title('Model ' + metric.title() + ':')
        ax.set_xlabel('Epoch')
        ax.set_ylabel(metric.title())
        ax.legend(['Train', 'Valid'], loc='upper left')
```

```python
plot_curve(H, valid=True)
```

# LeNet 5 (Convolution)

```python
def lenet5(verbose=True):
    model = keras.Sequential(name='LeNet5')
    # 6  16
    model.add(layers.Conv2D(filters=25, kernel_size=(5, 5), activation='relu', input_shape=(28,28,1), name='conv1_3x3', padding='valid'))
    model.add(layers.AveragePooling2D(name='pool1_2x2'))

    model.add(layers.Conv2D(filters=50, kernel_size=(5, 5), activation='relu', name='conv2_3x3'))
    model.add(layers.AveragePooling2D(name='pool2_2x2'))

    model.add(layers.Flatten())

    model.add(layers.Dense(units=120, activation='relu'))
    
    model.add(layers.Dense(units=84, activation='relu'))

    model.add(layers.Dense(units=10, activation = 'softmax'))
    if verbose: model.summary()
    return model
```

```python
lenet = lenet5()
```

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
x_train.shape, x_test.shape
```

```python
lenet.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
H_lenet = lenet.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=32, epochs=5)
```

```python
plot_curve(H_lenet, valid=True)
```

```python
from sklearn.metrics import log_loss
from sklearn.preprocessing import OneHotEncoder
```

```python
def most_confused_(n, model, valid_data, loss_func, return_images=True):
    x_true, y_true = valid_data[0], valid_data[1].copy()
    y_pred = model.predict(x_true)
    label = OneHotEncoder()
    y_true = y_true[:, None]
    y_true = label.fit_transform(y_true)
    losses = loss_func(y_true, y_pred)
    sort = np.argsort(losses)
    sort = sort[-n:]
    y_pred = model.predict_classes(x_true)
    y_true = valid_data[1]
    if return_images: return x_true[sort], y_true[sort], y_pred[sort], losses[sort]
    return sort, y_pred[sort], losses[sort]
```

```python
def cross_entropy(y_true, y_pred):
    return -np.array([t.dot(np.log(1/p))[0] for t, p in zip(y_true, y_pred)])
```

```python
def show_image(img, t_label, p_label, loss, ax=None, figsize=None):
    if ax is None: fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(img)
    ax.set_title('true: ' + class_names[t_label] + ', predicted: ' + class_names[p_label] + ', loss: ' + str(loss))

def show_batch(nrows, ncols, images, t_labels, p_labels, loss, figsize:(int, int)=(5, 5)):
    fig, axs = plt.subplots(nrows, ncols, figsize=(figsize[0]*nrows, figsize[1]*ncols))
    k = 0
    for i in range(nrows):
        for j in range(ncols):
            show_image(images[k], t_labels[k], p_labels[k], loss[k], axs[i, j])
            k += 1
```

```python
def plot_most_confused(n, model, valid_data, most_confused=None):
    if most_confused is None:
        x, t, p, l = most_confused_(n, model, valid_data, cross_entropy)
        print(l)
    else:
        x, t, p, l = most_confused
    nrows, ncols = int(np.ceil(n//3)), 3
    show_batch(nrows, ncols, x, t, p, l)
```
